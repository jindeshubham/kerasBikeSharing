```python
from keras.models import Sequential
import numpy as np
import pandas as pd 
from keras.layers import Activation, Dense, Dropout

train = pd.read_csv("train.csv")

categorical_columns = ["season","holiday","workingday","weather"]

categorical_dummy = []
dummy = {}
for each in categorical_columns:
    dummy = pd.get_dummies(train[each],prefix=each, drop_first=False)
    train = pd.concat([train, dummy], axis=1)
train.drop(categorical_columns,axis=1,inplace=True)
quant_features = ["windspeed","humidity","windspeed","casual","registered","count"]
scaled_features = {}
for each in quant_features:
    mean , std = train[each].mean() , train[each].std()
    train.loc[:,each] = (train[each] - mean)/std
time = []
for each in train["datetime"]:
     time.append(each.split(" ")[1].split(":")[0])
train["time"] = time
train.drop(columns=["datetime"],inplace=True)
print (train.head)

```

```python
test_data = train[-21*24:]

# Now remove the test data from the data set 
data = train[:-21*24]
target_fields = ['count', 'casual', 'registered']
features, targets = data.drop(target_fields, axis=1), data[target_fields]
test_features, test_targets = test_data.drop(target_fields, axis=1), test_data[target_fields]
```

```python
train_features, train_targets = features[:-60*24], targets[:-60*24]
val_features, val_targets = features[-60*24:], targets[-60*24:]
```

```python
model = Sequential()

model.add(Dense(units=32,activation='relu',input_dim = 17))
model.add(Dropout(0.2))
model.add(Dense(units=64,activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units=3))
model.compile(loss='mean_squared_error',
              optimizer='sgd',
              metrics=['accuracy'])
model.fit(train_features, train_targets, epochs=200, batch_size=32)
loss_and_metrics = model.evaluate(val_features, val_targets, batch_size=128)
```
